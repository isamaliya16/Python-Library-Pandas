```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

class SalesDataAnalyzer:
    def __init__(self, file_path=None):
        self.data = None
        if file_path:
            self.load_data(file_path)
        print("Sales Data Analyzer Initialized.")

    def __del__(self):
        print("Exiting Sales Data Analyzer. Goodbye!")

    def load_data(self, file_path):
        try:
            self.data = pd.read_csv(file_path)
            print("Dataset loaded successfully!")
        except Exception as e:
            print("Error loading dataset:", e)

    # Explore dataset
    def explore_data(self):
        if self.data is not None:
            print("\n1. First 5 rows:\n", self.data.head())
            print("\n2. Last 5 rows:\n", self.data.tail())
            print("\n3. Column names:\n", list(self.data.columns))
            print("\n4. Data types:\n", self.data.dtypes)
            print("\n5. Basic info:\n")
            print(self.data.info())
        else:
            print("Load dataset first!")

    # Handle missing values
    def clean_data(self):
        if self.data is not None:
            print("\n1. Display rows with missing values")
            print("2. Fill missing values with mean")
            print("3. Drop rows with missing values")
            print("4. Replace missing values with a specific value")
            choice = input("Enter your choice: ")
            if choice == '1':
                print(self.data[self.data.isnull().any(axis=1)])
            elif choice == '2':
                self.data.fillna(self.data.mean(numeric_only=True), inplace=True)
                print("Missing values filled with mean.")
            elif choice == '3':
                self.data.dropna(inplace=True)
                print("Rows with missing values dropped.")
            elif choice == '4':
                val = input("Enter value to replace missing data: ")
                self.data.fillna(val, inplace=True)
                print("Missing values replaced.")
        else:
            print("Load dataset first!")

    # Mathematical operations
    def mathematical_operations(self):
        if self.data is not None:
            print("\nSum of numeric columns:\n", self.data.sum(numeric_only=True))
            print("\nMean of numeric columns:\n", self.data.mean(numeric_only=True))
            print("\nMax values:\n", self.data.max(numeric_only=True))
            print("\nMin values:\n", self.data.min(numeric_only=True))
        else:
            print("Load dataset first!")

    # Search, sort, and filter
    def search_sort_filter(self):
        if self.data is not None:
            print("\n1. Search Product")
            print("2. Sort by Sales")
            print("3. Filter by Region")
            choice = input("Enter your choice: ")
            if choice == '1':
                product = input("Enter product name to search: ")
                print(self.data[self.data['Product'].str.contains(product, case=False, na=False)])
            elif choice == '2':
                print(self.data.sort_values(by='Sales', ascending=False))
            elif choice == '3':
                region = input("Enter region name: ")
                print(self.data[self.data['Region'].str.contains(region, case=False, na=False)])
        else:
            print("Load dataset first!")

    # Aggregating & statistical functions
    def aggregate_functions(self):
        if self.data is not None:
            print("\nTotal Sales:", self.data['Sales'].sum())
            print("Average Profit:", self.data['Profit'].mean())
            print("\nDescriptive Statistics:\n", self.data.describe())
        else:
            print("Load dataset first!")

    # Pivot Table
    def create_pivot_table(self):
        if self.data is not None:
            table = pd.pivot_table(self.data, values='Sales', index='Region', columns='Product', aggfunc='sum', fill_value=0)
            print("\nPivot Table:\n", table)
        else:
            print("Load dataset first!")

    # Advanced Operations
    def advanced_operations(self):
        if self.data is not None:
            print("\n=== Advanced Operations ===")
            print("1. Create Pivot Table (Sales by Region and Product)")
            print("2. Reindex & Rename Columns")
            print("3. GroupBy and Transform Operations")
            choice = input("Enter your choice: ")

            if choice == '1':
                pivot = pd.pivot_table(self.data, values='Sales', index='Region', columns='Product', aggfunc='sum', fill_value=0)
                print("\nPivot Table:\n", pivot)

            elif choice == '2':
                print("\nBefore Reindexing:\n", self.data.head())
                self.data = self.data.reset_index(drop=True)
                self.data.rename(columns={'Sales': 'Total_Sales', 'Profit': 'Total_Profit'}, inplace=True)
                print("\nAfter Reindexing & Renaming Columns:\n", self.data.head())

            elif choice == '3':
                if 'Region' in self.data.columns and 'Sales' in self.data.columns:
                    grouped = self.data.groupby('Region')['Sales'].transform('mean')
                    self.data['Avg_Sales_By_Region'] = grouped
                    print("\nAdded 'Avg_Sales_By_Region' column using groupby and transform:")
                    print(self.data[['Region', 'Sales', 'Avg_Sales_By_Region']].head())
                else:
                    print("Columns 'Region' or 'Sales' not found!")
        else:
            print("Load dataset first!")

    # Data Visualization
    def visualize_data(self):
        if self.data is not None:
            print("\n1. Bar Plot")
            print("2. Line Plot")
            print("3. Scatter Plot")
            print("4. Pie Chart")
            print("5. Histogram")
            print("6. Heatmap")
            choice = input("Enter your choice: ")

            if choice == '1':
                sns.barplot(x='Product', y='Sales', data=self.data)
                plt.title("Sales by Product")
            elif choice == '2':
                sns.lineplot(x='Year', y='Sales', data=self.data)
                plt.title("Sales over Years")
            elif choice == '3':
                x = input("Enter x-axis column: ")
                y = input("Enter y-axis column: ")
                sns.scatterplot(x=x, y=y, data=self.data)
                plt.title(f"{y} vs {x}")
            elif choice == '4':
                if 'Region' in self.data.columns:
                    self.data.groupby('Region')['Sales'].sum().plot(kind='pie', autopct='%1.1f%%')
                    plt.title("Sales by Region")
                else:
                    print(" 'Region' column not found!")
            elif choice == '5':
                self.data['Sales'].plot(kind='hist', bins=10, color='skyblue')
                plt.title("Sales Distribution")
            elif choice == '6':
                sns.heatmap(self.data.corr(numeric_only=True), annot=True, cmap='coolwarm')
                plt.title("Heatmap of Correlations")

            plt.xlabel("")
            plt.ylabel("")
            plt.show()
        else:
            print("Load dataset first!")

    # Save Visualization
    def save_visualization(self):
        filename = input("Enter file name to save (e.g. chart.png): ")
        plt.savefig(filename)
        print(f"Visualization saved as {filename}")


# Main Menu
def menu():
    analyzer = None
    while True:
        print("\n========== Data Analysis & Visualization Program ==========")
        print("1. Load Dataset")
        print("2. Explore Data")
        print("3. Handle Missing Data")
        print("4. Perform Mathematical Operations")
        print("5. Search/Sort/Filter")
        print("6. Aggregation & Statistics")
        print("7. Create Pivot Table")
        print("8. Data Visualization")
        print("9. Advanced Operations")
        print("10. Save Visualization")
        print("11. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            file_path = input("Enter CSV file path (e.g. sales_data.csv): ")
            analyzer = SalesDataAnalyzer(file_path)
        elif choice == '2' and analyzer:
            analyzer.explore_data()
        elif choice == '3' and analyzer:
            analyzer.clean_data()
        elif choice == '4' and analyzer:
            analyzer.mathematical_operations()
        elif choice == '5' and analyzer:
            analyzer.search_sort_filter()
        elif choice == '6' and analyzer:
            analyzer.aggregate_functions()
        elif choice == '7' and analyzer:
            analyzer.create_pivot_table()
        elif choice == '8' and analyzer:
            analyzer.visualize_data()
        elif choice == '9' and analyzer:
            analyzer.advanced_operations()
        elif choice == '10' and analyzer:
            analyzer.save_visualization()
        elif choice == '11':
            print("Exiting Program. Thank you!")
            break
        else:
            print("Invalid choice or load dataset first.")

if __name__ == "__main__":
    menu()
```
